The dataset contains information about restaurants listed on Zomato.
It includes restaurant names, locations, ratings, cuisines, and pricing information.
The data covers multiple cities and countries.
The dataset helps analyze restaurant trends and customer preferences.
It can be used to understand factors affecting restaurant ratings.

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

1. Missing values were removed using dropna().
2. Duplicate rows were removed.
3. Dataset consistency was checked using info().
4. Data types were verified.
5. Cleaned dataset was used for analysis.

```python
df['City'].value_counts().head(10)
```

```python
df['Cuisines'].value_counts().head(10)
```

```python
df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
```

```python
df['Has Online delivery'].value_counts()
```

```python
df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False).head(10)
```

```python
df['City'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Cities by Restaurant Count")
plt.xlabel("City")
plt.ylabel("Restaurants")
plt.show()
```

```python
plt.hist(df['Aggregate rating'])

plt.title("Restaurant Ratings Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Has Online delivery'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Online Delivery Availability")
plt.ylabel("")
plt.show()
```

```python
city_rating = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

city_rating.plot(kind='line')

plt.title("Average Rating by City")
plt.xlabel("City")
plt.ylabel("Rating")
plt.show()
```

```python
plt.scatter(
    df['Average Cost for two'],
    df['Aggregate rating']
)

plt.title("Cost vs Rating")
plt.xlabel("Average Cost for Two")
plt.ylabel("Rating")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

1. Certain cities have significantly more restaurants than others.
2. Some cuisines dominate the restaurant market.
3. Restaurants with online delivery are common in major cities.
4. Higher restaurant costs do not always guarantee higher ratings.
5. Customer ratings vary considerably across locations.

The most surprising finding was that expensive restaurants did not always receive the highest ratings. Several moderately priced restaurants performed equally well or better.
